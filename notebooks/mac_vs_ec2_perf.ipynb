```python
import pandas as pd
# mac_scores = pd.read_csv("outputs/metrics/model_comparison_test.csv")
# ec2_scores = pd.read_csv("outputs/metrics/model_comparison_test_ec2.csv")
# mac_lat = pd.read_csv("outputs/metrics/engine_benchmark_summary_test.csv")
# ec2_lat = pd.read_csv("outputs/metrics/engine_benchmark_summary_test_ec2.csv")

mac_scores = pd.read_csv("../outputs/metrics/model_comparison_test.csv")
ec2_scores = pd.read_csv("../outputs/metrics/model_comparison_test_ec2.csv")
mac_lat = pd.read_csv("../outputs/metrics/engine_benchmark_summary_test.csv")
ec2_lat = pd.read_csv("../outputs/metrics/engine_benchmark_summary_test_ec2.csv")

mac_scores["machine"] = "mac"
ec2_scores["machine"] = "ec2"
scores = pd.concat([mac_scores, ec2_scores], ignore_index=True)
mac_lat["machine"] = "mac"
ec2_lat["machine"] = "ec2"
lat = pd.concat([mac_lat, ec2_lat], ignore_index=True)
# Keep high-signal columns
scores_cols = [c for c in [
    "machine","backbone",
    "class_accuracy","gender_accuracy","color_accuracy","usage_accuracy","joint_accuracy",
    "class_macro_f1","gender_macro_f1","color_macro_f1","usage_macro_f1"
] if c in scores.columns]
lat_cols = [c for c in [
    "machine","backbone","batch_size","onnx_static_batch_1",
    "pytorch_mean_ms_per_image","pytorch_p95_ms_per_image",
    "onnx_mean_ms_per_image","onnx_p95_ms_per_image",
    "speedup_pytorch_over_onnx"
] if c in lat.columns]
out_scores = "../outputs/metrics/mac_vs_ec2_scores.csv"
out_lat = "../outputs/metrics/mac_vs_ec2_latency.csv"
scores[scores_cols].sort_values(["backbone","machine"]).to_csv(out_scores, index=False)
lat[lat_cols].sort_values(["backbone","machine"]).to_csv(out_lat, index=False)
print("Wrote:", out_scores)
print("Wrote:", out_lat)
```
